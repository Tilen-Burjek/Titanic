```python
#importing all the libraries that I'll most likely use
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

import pandas as pd
from pandas import Series,DataFrame

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#the data is already separated into a train and test set
titanic_train = pd.read_csv("data//train.csv")
titanic_test  = pd.read_csv("data//test.csv")
titanic_full = (titanic_train, titanic_test)
```

Alright let's look at the values individually and decide which ones to drop, how to deal with missing values, etc


```python
#Let's check which attributes have missing values
titanic_train.isna().any()
```

```python
#Let's start with the obvious one. We can clearly see that the gender of the passenger greatly impacted their chance of survival
titanic_train[['Sex', 'Survived']].groupby(['Sex'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
#The other obvious one is class. It is clear that passenger from a higher class are more likely to survive than from lower classes
titanic_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
#The line below fetches too many results so I'll categorize the age in ranges
#titanic_train[['Age', 'Survived']].groupby(['Age'], as_index=False).mean().sort_values(by='Survived', ascending=False)
graph = sns.FacetGrid(titanic_train, col='Survived')
graph.map(plt.hist, 'Age', bins=30)
```

I thought about removing the title but someone on Kaggle had the idea to unify and modify it a bit. I'll leave the ones which occur the most and replace others with 'Rare'

```python
for ds in titanic_full:
    ds['Title'] = ds.Name.str.extract(' ([A-Za-z]+)\.', expand=False)

pd.crosstab(titanic_train['Title'], titanic_train['Sex'].sort_values(ascending=True))
```

```python
for dataset in titanic_full:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    
titanic_train[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

```python
mapps = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5, "Mlle": 2, "Mme": 1, "Ms": 2}
for dataset in titanic_full:
    dataset['Title'] = dataset['Title'].map(mapps)
    #dataset['Title'] = dataset['Title'].fillna(0)

titanic_train.head()
```

```python
#After we've got the title from the name it is no longer of any importance to us
titanic_train = titanic_train.drop(['Name', 'PassengerId'], axis=1)
titanic_test = titanic_test.drop(['Name'], axis=1)
titanic_full = [titanic_train, titanic_test]
titanic_train.shape, titanic_test.shape
```

```python
titanic_train.head()
```

```python
#Let's categorize all attributes
titanic_train['Sex'] = titanic_train['Sex'].map( {'female': 1, 'male': 0} )
titanic_test['Sex'] = titanic_test['Sex'].map( {'female': 1, 'male': 0} )
titanic_train.head()
```

```python
#Categorize the ages
titanic_train.loc[ titanic_train['Age'] <= 16, 'Age'] = 0
titanic_train.loc[(titanic_train['Age'] > 16) & (titanic_train['Age'] <= 32), 'Age'] = 1
titanic_train.loc[(titanic_train['Age'] > 32) & (titanic_train['Age'] <= 48), 'Age'] = 2
titanic_train.loc[(titanic_train['Age'] > 48) & (titanic_train['Age'] <= 64), 'Age'] = 3
titanic_train.loc[ titanic_train['Age'] > 64, 'Age']
titanic_train.head()
```

```python
titanic_test.loc[ titanic_test['Age'] <= 16, 'Age'] = 0
titanic_test.loc[(titanic_test['Age'] > 16) & (titanic_test['Age'] <= 32), 'Age'] = 1
titanic_test.loc[(titanic_test['Age'] > 32) & (titanic_test['Age'] <= 48), 'Age'] = 2
titanic_test.loc[(titanic_test['Age'] > 48) & (titanic_test['Age'] <= 64), 'Age'] = 3
titanic_test.loc[ titanic_test['Age'] > 64, 'Age']
#titanic_test.head()
```

```python
#Create family size by mergin SibSp (siblings) and Parch (parent/child)
titanic_test['FamilySize'] = titanic_test['SibSp'] + titanic_test['Parch'] + 1
titanic_train['FamilySize'] = titanic_train['SibSp'] + titanic_train['Parch'] + 1
```

```python
#Create attribute IsAlone by checking family size
titanic_test['IsAlone'] = 0
titanic_test.loc[titanic_test['FamilySize'] == 1, 'IsAlone'] = 1

titanic_train['IsAlone'] = 0
titanic_train.loc[titanic_train['FamilySize'] == 1, 'IsAlone'] = 1
titanic_train[['IsAlone', 'Survived']].groupby(['IsAlone'], as_index=False).mean()
```

```python
titanic_train = titanic_train.drop(['Parch', 'SibSp', 'FamilySize'], axis=1)
titanic_test = titanic_test.drop(['Parch', 'SibSp', 'FamilySize'], axis=1)
titanic_full = [titanic_train, titanic_test]
```

```python
#Change missing value in embarked with most frequent one
most_freq = titanic_train.Embarked.dropna().mode()[0]
titanic_train['Embarked'] = titanic_train['Embarked'].fillna(most_freq)
titanic_test['Embarked'] = titanic_test['Embarked'].fillna(most_freq)
titanic_train[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
#change to numerical
titanic_train['Embarked'] = titanic_train['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
titanic_test['Embarked'] = titanic_test['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
```

```python
titanic_train.head()
```

```python
#Let's decide where to split the values
titanic_train['FareBand'] = pd.qcut(titanic_train['Fare'], 4)
titanic_train[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=True)
```

```python
#Categorize the fare values
titanic_train.loc[ titanic_train['Fare'] <= 7.91, 'Fare'] = 0
titanic_train.loc[(titanic_train['Fare'] > 7.91) & (titanic_train['Fare'] <= 14.454), 'Fare'] = 1
titanic_train.loc[(titanic_train['Fare'] > 14.454) & (titanic_train['Fare'] <= 31), 'Fare']   = 2
titanic_train.loc[ titanic_train['Fare'] > 31, 'Fare'] = 3
#titanic_train['Fare'] = titanic_train['Fare'].astype(int)

titanic_test.loc[ titanic_test['Fare'] <= 7.91, 'Fare'] = 0
titanic_test.loc[(titanic_test['Fare'] > 7.91) & (titanic_test['Fare'] <= 14.454), 'Fare'] = 1
titanic_test.loc[(titanic_test['Fare'] > 14.454) & (titanic_test['Fare'] <= 31), 'Fare']   = 2
titanic_test.loc[ titanic_test['Fare'] > 31, 'Fare'] = 3
#titanic_test['Fare'] = titanic_test['Fare'].astype(int)

titanic_full = (titanic_train, titanic_test)
```

```python
#titanic_test.head(10)
```

```python
#Remove the attributes that we do not need
titanic_train = titanic_train.drop(['FareBand', 'Ticket', 'Cabin'], axis=1)
titanic_test    = titanic_test.drop(['Ticket', 'Cabin'], axis=1)
#titanic_test['Age'] = titanic['Age']
```

```python
titanic_test = titanic_test[pd.notnull(titanic_test['Age'])]
titanic_train = titanic_train[pd.notnull(titanic_train['Age'])]
titanic_test['Fare'] = titanic_test['Fare'].fillna(0)
```

```python
titanic_train.head(10)
```

```python
titanic_test.head(10)
```

```python
X_train = titanic_train.drop("Survived", axis=1)
Y_train = titanic_train["Survived"]
X_test  = titanic_test.drop("PassengerId", axis=1).copy()
```

```python
X_train.shape, Y_train.shape, X_test.shape
```

```python
svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
acc_svc
```

```python
# Logistic Regression

logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
coeff_df = pd.DataFrame(titanic_train.columns.delete(0))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(logreg.coef_[0])

coeff_df.sort_values(by='Correlation', ascending=False)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn
```

```python
# Gaussian Naive Bayes

gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
Y_pred = gaussian.predict(X_test)
acc_gaussian = round(gaussian.score(X_train, Y_train) * 100, 2)
acc_gaussian
```

```python
# Perceptron

perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
Y_pred = perceptron.predict(X_test)
acc_perceptron = round(perceptron.score(X_train, Y_train) * 100, 2)
acc_perceptron
```

```python
# Linear SVC

linear_svc = LinearSVC()
linear_svc.fit(X_train, Y_train)
Y_pred = linear_svc.predict(X_test)
acc_linear_svc = round(linear_svc.score(X_train, Y_train) * 100, 2)
acc_linear_svc
```

```python
# Stochastic Gradient Descent

sgd = SGDClassifier()
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)
acc_sgd = round(sgd.score(X_train, Y_train) * 100, 2)
acc_sgd
```

```python
# Decision Tree

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
acc_decision_tree
```

```python
# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Perceptron', 
              'Stochastic Gradient Decent', 'Linear SVC', 
              'Decision Tree'],
    'Score': [acc_svc, acc_knn, acc_log, 
              acc_random_forest, acc_gaussian, acc_perceptron, 
              acc_sgd, acc_linear_svc, acc_decision_tree]})
models.sort_values(by='Score', ascending=False)
```
